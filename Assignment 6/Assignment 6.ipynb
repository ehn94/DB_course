As said a b-tree is god for searching. In this case its great because it can be sorted alphabetically. 

idx_17102_url
This is an index of type GiST. 
GiST is good to use for two-dimensional geometric data types. 
As we can see, position is of type point, which is a data type containing (x,y). 
GiST indexes can also be used for optimizing "nearest-neighbor" searches, which could be relevant in this case. 
This index has a size of 8192 byte.

```python
%sql CREATE MATERIALIZED VIEW race_winners_caches AS SELECT races.date, surname, dob, milliseconds, races.name racename, circuits.name FROM results JOIN races USING(raceid) JOIN drivers USING (driverid)JOIN circuits USING(circuitid) WHERE dob > '1980-01-01' AND rank = 1;
```

```python
%sql EXPLAIN ANALYZE SELECT * from race_winners_caches
```

With MATERIALIZED VIEW we make a view which wil be cached. 
This is why the execution time is so short- because it have been cached, and it won't have to execute the whole query again.
